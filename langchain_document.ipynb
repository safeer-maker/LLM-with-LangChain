```python
import os
from langchain.document_loaders import PyPDFLoader 
```

```python
data_path = 'data'
files_name = os.listdir(data_path)
files_name
```

```python
loader = []

for file in files_name:
    if (file[-3:] == 'pdf'):
        path = os.path.join (data_path, file)
        loader.extend ( PyPDFLoader(path).load() )

loader
```

# lets have question answer with our data

```python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.retrievers import ContextualCompressionRetriever
from langchain_core.prompts import ChatPromptTemplate
from langchain.retrievers.document_compressors import LLMChainExtractor

model = "gpt-3.5-turbo-0125"
db_dir = "data/chroma_db"
```

```python
llm = ChatOpenAI (model=model)
embeddings = OpenAIEmbeddings()
chroma_db = Chroma (
                embedding_function=embeddings,
                persist_directory= db_dir
                )

compressor = LLMChainExtractor.from_llm (llm=llm)
compressor
```

```python
context_retriver = ContextualCompressionRetriever (
    base_compressor=compressor , 
    base_retriever= chroma_db.as_retriever()
)

context_retriver
```

```python
question = "what are the facts about increase in e-commerce barket and how big is over target audiance?"
```

```python
def context_creator (inp):
    context = str()
    for i in inp:
        context = context + '*' * 20 + " \n\n "
        context = context + i.page_content 
    context = context + "*" * 50

    return context

# x = context_creator(['asba', 'asfasdf','fsgdfg'])
# x 
```

```python
chroma_db.max_marginal_relevance_search (question)
```

```python
context = context_retriver.invoke(question)
context
```

```python
context_ord = context_creator (context)
context_ord
```

```python
prompt = """
Your are a question.
Context is provided to you in context section.
If you dont know the answer write you do not know about it.

Context : {context}

question : {question}

answer :
"""

template = ChatPromptTemplate.from_template(prompt)
template
```

```python
chain = template | llm
chain
```

```python
ans  =  chain.invoke (
    {"context": context_ord, "question": question}
)
ans
```

```python
ans.content
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

chain_rag = (
    {"context": context_retriver | format_docs, "question": RunnablePassthrough()}
    | template #prompt
    | llm
    | StrOutputParser()
)
```

```python
ans_rag = chain_rag.invoke (question)
ans_rag
```
