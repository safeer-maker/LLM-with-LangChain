```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
import os
import chromadb
```

```python
dir_path = "data"

files_list = os.listdir(dir_path)
files_list

pdf_files = [dir_path + "/" + files for files in files_list if files[-3:] == 'pdf']
pdf_files
```

```python
PyPDFLoader( 'data/41466639102.pdf')
```

```python
raw_pdf_data = []
raw_pdf_data.extend( PyPDFLoader (file) for file in pdf_files )
pages =  raw_pdf_data[0].load()
```

```python
pages, len(pages)
```

```python
spliter = RecursiveCharacterTextSplitter (
            chunk_size = 250,
            chunk_overlap = 20,
        )

spliter
```

```python
pages_250 = spliter.split_documents (pages)
pages_250, len (pages_250)
```

```python
pages_250[0].metadata
```

```python
pages_250[0].metadata['source']
```

```python
ids=[ str (i) for i in range (len( pages_250 )) ]
documents= [doc.page_content for doc in pages_250]
metadatas= [{'source': f" {doc.metadata['source']} and page {doc.metadata['page']}"} for doc in pages_250]
```

```python
metadatas
```

```python
documents
```

```python
collection.add (
    ids= ids,
    documents= documents,
    metadatas= metadatas,
    # embeddings=embedded_fun
)
```
